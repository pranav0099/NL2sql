```python
# Cell 1 — Setup
import sys
sys.path.insert(0, r"C:\Users\prana\OneDrive\Desktop\projects\ns")
import torch
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
```

```python
# Cell 2 — Load vocab and data
from dl.dataset import load_vocab, get_dataloaders

nl_vocab, sql_vocab = load_vocab("data/processed/vocab.json")
print(f"NL vocab size  : {nl_vocab['size']}")
print(f"SQL vocab size : {sql_vocab['size']}")

train_loader, val_loader = get_dataloaders(
    "data/processed/train.json",
    "data/processed/val.json",
    nl_vocab, sql_vocab,
    batch_size=64        # larger batch = faster training
)
print(f"Train batches : {len(train_loader)}")
print(f"Val batches   : {len(val_loader)}")
```

```python
# Cell 3 — Create model
from dl.model import create_model
model = create_model(nl_vocab, sql_vocab)
print(f"Parameters: {model.count_parameters():,}")
```

```python
# Cell 4 — Train with progress bar
from dl.trainer import TransformerTrainer
from tqdm.notebook import tqdm   # nicer progress bar in Jupyter

trainer = TransformerTrainer(
    model         = model,
    nl_vocab      = nl_vocab,
    sql_vocab     = sql_vocab,
    learning_rate = 1e-4
)

# Change num_epochs based on your patience:
# 5  epochs = ~15 min CPU  | good for testing
# 15 epochs = ~45 min CPU  | good for demo
# 30 epochs = ~3 hrs CPU   | best quality

history = trainer.train(
    train_loader = train_loader,
    val_loader   = val_loader,
    num_epochs   = 15,           # change this number
    save_path    = "models/saved/transformer_nl2sql.pt"
)
```

```python
# Cell 5 — Plot training curve
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(history["train_losses"], label="Train Loss", color="blue")
plt.plot(history["val_losses"],   label="Val Loss",   color="orange")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("NL2SQL Transformer Training")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("logs/training_curve.png")
plt.show()
print(f"Best val loss : {history['best_val_loss']:.4f}")
print(f"Best epoch    : {history['best_epoch']}")
```

```python
# Cell 6 — Test generator
from dl.generator import SQLGenerator

generator = SQLGenerator(
    model_path = "models/saved/transformer_nl2sql.pt",
    vocab_path = "data/processed/vocab.json",
    db_path    = "database/sample.db"
)

TEST_QUERIES = [
    "Show all customers from Mumbai",
    "What is the average salary of employees",
    "Count total orders by city",
    "Top 5 products by price",
    "Show customer names with their orders",
    "List employees with salary above 60000",
    "Total sales amount for each category",
    "Show most recent 10 orders",
]

print("="*65)
print("SQL Generator — Test Results")
print("="*65)
for i, query in enumerate(TEST_QUERIES, 1):
    result = generator.generate(query)
    print(f"\n[{i}] Question : {query}")
    print(f"    SQL      : {result['sql']}")
    print(f"    Conf     : {result['confidence']:.0%}")
```

```python
# Cell 7 — BLEU Score evaluation
bleu = generator.evaluate_bleu(
    test_path   = "data/processed/test.json",
    max_samples = 500
)
print(f"BLEU Score   : {bleu['bleu_score']:.4f}")
print(f"Exact Match  : {bleu['exact_match']:.1%}")
print(f"Avg Conf     : {bleu['avg_confidence']:.1%}")
print(f"Total Tested : {bleu['total_samples']}")
```
